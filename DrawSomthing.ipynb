```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(start=-10, stop=10, step=0.1)
```

```python
plt.plot(x, np.tan(x))
plt.title("y = tan(x)")
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'red')
plt.plot(x, np.cos(x), color = 'blue')
plt.title("Sin and Cos")
plt.show()
```

```python
plt.plot(x, np.exp(x), color = 'red')
plt.title("Sin and Cos")
plt.show()
```

```python
plt.plot(x, 1/(1+np.exp(-x)))
plt.title("sigmoid(x)")
plt.show()
```
