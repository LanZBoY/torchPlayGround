# 使用優化器來調整參數

除了用梯度相減的方法之外，另外還有使用優化器來實現梯度下降
給定一筆資料y跟x以及一個f()
找到一個f(x)來逼近y值

```python
import torch
```

```python
y = torch.tensor([10.,20.,30.], dtype=torch.float32)
x = torch.randn(size=(6,))
w = torch.randn(size=(3,6), requires_grad=True)
b = torch.randn(size=(3,), requires_grad=True)
```

```python
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(params=[w, b], lr=1e-2)
```

```python
for epoch in range(200):
    optimizer.zero_grad()
    z = w @ x + b
    loss = loss_fn(y, z)
    print(f'Loss = {loss.item()}')
    print(f' y = {y}')
    print(f' z = {z}')
    loss.backward()
    optimizer.step()

print(f'Final x is = {x}')
```
