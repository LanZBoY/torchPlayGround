```python
import torch
import numpy as np
from NLPUtils.DataModel import Vocabulary
from torch.utils.data import DataLoader, TensorDataset
import pickle
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
import seaborn as sns
```

```python
with open('./SLTCDataset/mr/train_voc.data', 'rb') as f:
    train_voc : Vocabulary = pickle.load(f)
x_train = np.load('./SLTCDataset/mr/x_train.npy')
y_train = np.load('./SLTCDataset/mr/y_train.npy')
x_test = np.load('./SLTCDataset/mr/x_test.npy')
y_test = np.load('./SLTCDataset/mr/y_test.npy')
```

```python
trainDataset = TensorDataset(torch.from_numpy(x_train), torch.from_numpy(y_train))
testDataset = TensorDataset(torch.from_numpy(x_test), torch.from_numpy(y_test))
```

```python
trainDataLoader = DataLoader(trainDataset, batch_size = 32, shuffle=True)
```

```python
it = iter(trainDataLoader)
bat_x, bat_y = next(it)
```

```python
class EMB_MLP(torch.nn.Module):

    def __init__(self, train_voc : Vocabulary) -> None:
        super(EMB_MLP, self).__init__()
        self.embedding = torch.nn.EmbeddingBag(train_voc.num_words, embedding_dim=300, padding_idx=0)
        self.linear_d = torch.nn.Linear(300, 128)
        self.linear_o = torch.nn.Linear(128, 2)
    
    def forward(self, x):
        h = self.embedding(x)
        h = self.linear_d(h)
        h = torch.nn.functional.relu(h)
        h = self.linear_o(h)
        return h
```

```python
emb_mlp = EMB_MLP(train_voc = train_voc).cuda()
loss_fn = torch.nn.CrossEntropyLoss()
otim = torch.optim.Adam(emb_mlp.parameters())
losses = []
```

```python
for ep in tqdm(range(20)):
    total_loss = 0.
    for x, y in trainDataLoader:
        x, y = x.cuda(), y.cuda()
        otim.zero_grad()
        pred = emb_mlp(x)
        loss = loss_fn(pred, y)
        loss.backward()
        otim.step()
        total_loss += loss.item()
    losses.append(total_loss / (len(testDataset) / 32))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(losses)
plt.title("Training Loss")
plt.show()
```

```python
with torch.no_grad():
    x, y = testDataset.tensors[0].cuda(), testDataset.tensors[1].cuda()
    pred = emb_mlp(x)
    count = (pred.argmax(1) == y).sum().float()
    print(f'Acc = {count / len(testDataset) * 100 : 0.3f}%')
```

```python
class LSTMBased(torch.nn.Module):

    def __init__(self, train_voc : Vocabulary) -> None:
        super(LSTMBased, self).__init__()
        self.embedding = torch.nn.Embedding(train_voc.num_words, 300, padding_idx = 0)
        self.lstm1 = torch.nn.LSTM(input_size = 300, hidden_size = 300, batch_first = True)
        self.linear_d = torch.nn.Linear(15 * 300, 128)
        self.linear_i = torch.nn.Linear(128, 2)
    def forward(self, x):
        h = self.embedding(x)
        h, (_, _) = self.lstm1(h)
        h = torch.nn.functional.relu(h)
        h = torch.flatten(h, start_dim = 1)
        h = self.linear_d(h)
        h = torch.nn.functional.relu(h)
        h = self.linear_i(h)
        return h
```

```python
lstm_based = LSTMBased(train_voc = train_voc).cuda()
loss_fn = torch.nn.CrossEntropyLoss()
otim = torch.optim.Adam(lstm_based.parameters())
losses = []
```

```python
for ep in tqdm(range(20)):
    total_loss = 0.
    for x, y in trainDataLoader:
        x, y = x.cuda(), y.cuda()
        otim.zero_grad()
        pred = lstm_based(x)
        loss = loss_fn(pred, y)
        loss.backward()
        otim.step()
        total_loss += loss.item()
    losses.append(total_loss / (len(testDataset) / 32))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(losses)
plt.title("Training Loss")
plt.show()
```

```python
with torch.no_grad():
    x, y = testDataset.tensors[0].cuda(), testDataset.tensors[1].cuda()
    pred = lstm_based(x)
    count = (pred.argmax(1) == y).sum().float()
    print(f'Acc = {count / len(testDataset) * 100 : 0.3f}%')
```

```python
class LSTMPredWithFinalState(torch.nn.Module):

    def __init__(self, train_voc : Vocabulary) -> None:
        super(LSTMPredWithFinalState, self).__init__()
        self.embedding = torch.nn.Embedding(train_voc.num_words, 300, padding_idx = 0)
        self.lstm1 = torch.nn.LSTM(input_size = 300, hidden_size = 300, batch_first = True)
        self.linear_d = torch.nn.Linear(300, 128)
        self.linear_i = torch.nn.Linear(128, 2)
    def forward(self, x):
        h = self.embedding(x)
        _, (h, _) = self.lstm1(h)
        h = torch.permute(h, (1, 0, 2))
        h = h.view(-1, 300)
        h = torch.nn.functional.relu(h)
        h = self.linear_d(h)
        h = torch.nn.functional.relu(h)
        h = self.linear_i(h)
        return h
```

```python
lstm_prd_fin = LSTMPredWithFinalState(train_voc = train_voc).cuda()
loss_fn = torch.nn.CrossEntropyLoss()
otim = torch.optim.Adam(lstm_prd_fin.parameters())
losses = []
```

```python
for ep in tqdm(range(20)):
    total_loss = 0.
    for x, y in trainDataLoader:
        x, y = x.cuda(), y.cuda()
        otim.zero_grad()
        pred = lstm_prd_fin(x)
        loss = loss_fn(pred, y)
        loss.backward()
        otim.step()
        total_loss += loss.item()
    losses.append(total_loss / (len(testDataset) / 32))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(losses)
plt.title("Training Loss")
plt.show()
```

```python
with torch.no_grad():
    x, y = testDataset.tensors[0].cuda(), testDataset.tensors[1].cuda()
    pred = lstm_prd_fin(x)
    count = (pred.argmax(1) == y).sum().float()
    print(f'Acc = {count / len(testDataset) * 100 : 0.3f}%')
```

```python
class AttentioBaseModel(torch.nn.Module):

    def __init__(self, train_voc : Vocabulary) -> None:
        super(AttentioBaseModel, self).__init__()
        self.embeding = torch.nn.Embedding(train_voc.num_words, embedding_dim = 300, padding_idx= 0)
        self.selfAtt = torch.nn.MultiheadAttention(300, 1, batch_first = True)
        self.mlp = torch.nn.Sequential(
            torch.nn.Linear(300, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 2)
        )
    def forward(self, x):
        e = self.embeding(x)
        h_s, atten_score = self.selfAtt(e, e, e)
        h_sum = h_s.sum(dim = 1)
        out = self.mlp(h_sum)
        return out, atten_score
```

```python
class AttentioBaseModelWithFirstHidden(torch.nn.Module):

    def __init__(self, train_voc : Vocabulary) -> None:
        super(AttentioBaseModelWithFirstHidden, self).__init__()
        self.embeding = torch.nn.Embedding(train_voc.num_words, embedding_dim = 300, padding_idx= 0)
        self.selfAtt = torch.nn.MultiheadAttention(300, 1, batch_first = True)
        self.mlp = torch.nn.Sequential(
            torch.nn.Linear(300, 128),
            torch.nn.ReLU(),
            torch.nn.Linear(128, 2)
        )
    def forward(self, x):
        e = self.embeding(x)
        h_s, atten_score = self.selfAtt(e, e, e)
        h_0 = h_s[:,0]
        out = self.mlp(h_0)
        return out, atten_score
```

```python
def drawHeatMap(index, atten_score, train_voc : Vocabulary):
    words = [train_voc.idx2word[i.item()] for i in testDataset.tensors[0][index]]
    label = "positive" if testDataset.tensors[1][index].item() == 1 else "negative"
    dataframe = pd.DataFrame(atten_score[index].cpu(), index=words, columns=words)
    plt.figure(figsize=(10, 10))
    sns.heatmap(data = dataframe, linewidths=0.5, vmin=0., vmax=1.)
    plt.title(label=label)
    plt.show()
```

```python
attentionModel = AttentioBaseModel(train_voc= train_voc).cuda()
loss_fn = torch.nn.CrossEntropyLoss()
otim = torch.optim.Adam(attentionModel.parameters())
losses = []
```

```python
for ep in tqdm(range(20)):
    total_loss = 0.
    for x, y in trainDataLoader:
        x, y = x.cuda(), y.cuda()
        otim.zero_grad()
        pred, _ = attentionModel(x)
        loss = loss_fn(pred, y)
        loss.backward()
        otim.step()
        total_loss += loss.item()
    losses.append(total_loss / (len(testDataset) / 32))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(losses)
plt.title("Training Loss")
plt.show()
```

```python
with torch.no_grad():
    x, y = testDataset.tensors[0].cuda(), testDataset.tensors[1].cuda()
    pred, atten_score = attentionModel(x)
    count = (pred.argmax(1) == y).sum().float()
    print(f'Acc = {count / len(testDataset) * 100 : 0.3f}%')
```

```python
drawHeatMap(index = 60, atten_score = atten_score, train_voc = train_voc)
```
