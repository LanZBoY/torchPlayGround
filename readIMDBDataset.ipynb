```python
import torch
import pandas as pd
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
```

```python
IMDB_file = './basic_dataset/IMDB/IMDB Dataset.csv'
df = pd.read_csv(IMDB_file)
```

```python
len(df)
```

```python
sentences = df.review
```

```python
tokenizer = get_tokenizer("basic_english")
```

```python
def yield_tokens(data_iter):
    for text in data_iter:
        yield tokenizer(text)
```

```python
def fixed_size (MAX_LEN = 100, tokens = []):
    if len(tokens) > MAX_LEN:
        return tokens[:MAX_LEN]
    while len(tokens) < MAX_LEN:
        tokens.append(0)
    return tokens
```

```python
voc = build_vocab_from_iterator(yield_tokens(sentences), specials=['<unk>'])
voc.set_default_index(voc['<unk>'])
```

```python
voc(tokenizer(sentences[0]))
```

```python
for s in sentences:
    print(len(fixed_size(MAX_LEN=300, tokens=voc(tokenizer(s)))))
    break
```
